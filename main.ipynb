```python
import os

files_to_skip = ['ucAdm.csv', 'ucDemoData.csv', 'ucRaceData.csv', '.DS_Store']
dfs = {}

for fname in os.listdir('data/oneYear'):
    if fname in files_to_skip:
        continue

    college = fname.split('D')[0]

    df = pd.read_csv(f'data/oneYear/{fname}', encoding='utf-16', delimiter='\t')
    df = df.rename(columns={'Unnamed: 0': 'Race', 'Unnamed: 1': 'Type'})
    df['Type'] = df['Type'].replace({'adm redacted calc along Characteristics FR': 'number', 'adm redacted percent calc along Characteristics FR': 'percent'})
    df_new = df.T.drop(columns=[0, 2, 4, 6, 8, 10, 12, 14])
    df_new = df_new.rename(columns={1: 'AfricanAmerican', 3: 'AmericanIndian', 5: 'Asian', 7: 'Hispanic/Latino(a)', 9: 'PacificIslander', 11: 'White',
                       13: 'Unknown', 15: 'International'})
    
    df_final = df_new.iloc[2::]
    df_final = df_final.replace(to_replace='%', value='', regex=True).astype(int)
    df_final['college'] = college
    df_final = df_final.reset_index().drop(columns=['index'])

    dfs[college] = df_final
```

```python
final_df = pd.concat(list(dfs.values())).reset_index(drop=True).drop(columns=['PacificIslander'])
final_df.to_csv('data/ucCollege.csv')
```

```python
files_to_skip = ['ucAdm.csv', 'ucDemoData.csv', 'ucRaceData.csv', '.DS_Store']
dfs = {}

for fname in os.listdir('data/allYears'):
    if fname in files_to_skip:
        continue

    college = fname.split('Years')[1][:-4]

    df = pd.read_csv(f'data/allYears/{fname}', encoding='utf-16', delimiter='\t')
    df = df.rename(columns={'Unnamed: 0': 'Race', 'Unnamed: 1': 'Type'})
    df['Type'] = df['Type'].replace({'adm redacted calc along Characteristics FR': 'number', 'adm redacted percent calc along Characteristics FR': 'percent'})
    df_new = df.T.drop(columns=[0, 2, 4, 6, 8, 10, 12, 14])
    df_new = df_new.rename(columns={1: 'AfricanAmerican', 3: 'AmericanIndian', 5: 'Asian', 7: 'Hispanic/Latino(a)', 9: 'PacificIslander', 11: 'White',
                       13: 'Unknown', 15: 'International'})
    
    df_final = df_new.iloc[2::].fillna(0)
    df_final = df_final.replace(to_replace='%', value='', regex=True).astype(int).drop(columns=['PacificIslander'])
    df_final = df_final.reset_index().rename(columns={'index': 'year'})

    dfs[college] = df_final

    df_final.to_csv(f'data/all{college}.csv')
```
